```python
import pandas as pd
```

```python
data = pd.read_csv("LB_TA.txt", delimiter = r"\s+")
data
```

```python
# Menghapus data yang duplikat
data = data.drop_duplicates()
data
```

```python
#  Menghapus data yang memiliki nilai kosong atau NaN pada kolom 'at'
data = data.dropna(subset=['at'])
data
```

```python
# Menghitung rata-rata nilai di kolom 'at' untuk nilai yang sama pada kolom 'tic'
data = data.groupby('tic')['at'].mean().reset_index()
data
```

```python
# Mengubah kolom 'at' menjadi 'avg_at'
data = data.rename(columns={'at': 'avg_at'})
data
```

```python
# Menyimpan hasil ke bentuk CSV
data.to_csv("Hasil Quiz.csv", index = False)

# Menampilkan hasilnya
data
```
